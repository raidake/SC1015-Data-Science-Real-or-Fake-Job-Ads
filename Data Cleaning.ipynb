# Data Cleaning Portion #


```python
#import libraries
import numpy as np
import pandas as pd
```

```python
jobdata = pd.read_csv("fake_job_postings.csv")
jobdata.head()
```

```python
print(jobdata.dtypes)
```

```python
jobdata.isnull().sum()
```

```python
job_pick_data = jobdata[['title','description','requirements','benefits','has_company_logo','required_experience',
                         'required_education','fraudulent']]
job_pick_data
```

```python
job_pick_data.isnull().sum()
```

```python
job_pick_data.fillna("None", inplace=True)
```

```python
job_pick_data.isnull().sum()
```

```python
job_pick_data.duplicated().sum()
```

```python
job_clean_data= job_pick_data.copy()
job_clean_data.drop_duplicates(inplace=True)
job_clean_data
```

```python
job_clean_data.to_csv("cleanJobData.csv",index=False)
```
